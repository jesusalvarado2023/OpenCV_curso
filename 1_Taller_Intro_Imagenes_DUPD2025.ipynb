# 3. Edición de píxeles

```python
# Los colores son Red 206, Green 170, Blue 70
# Podemos editar un pixel?
# Pintémoslo de negro

img1[200,0,:] = [0,0,0]

plt.imshow(img1, cmap="viridis")
plt.colorbar()
```

```python
# No es visible a simple vista
# Hacemos un zoom

zoom1 = img1[190:210, 0:20, :]
plt.imshow(zoom1)
```

```python
# Pintémoslo de azul (R,G,B=255)

img1[200,0,:] = [0,0,255]

zoom1 = img1[190:210, 0:20, :]
plt.imshow(zoom1)
```

```python
# Editemos un porcentaje de píxeles

Totpixeles = img1.shape[0]*img1.shape[1]
print("El total de píxeles es: ", Totpixeles)

PixelesPintar = int(0.05*Totpixeles)
print("El número de píxeles a pintar es: ", PixelesPintar)

# Generamos números aleatorios
import numpy as np
i = np.random.randint(0,1756)
j = np.random.randint(0,2000)

for rnd in range(PixelesPintar):
    i = np.random.randint(0,1756)
    j = np.random.randint(0,2000)
    img1[i,j,:] = [0,0,0]
plt.imshow(img1)
```
